```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

1. Remove any duplicate entries in the dataset. The PassengerId column can be used as a unique identifier for each passenger.



```python
#Loading the dataset
data = pd.read_csv('/content/titanic (1).csv')
```

```python
#Remove duplicate entries based on the 'PassengerId' column
df = data.drop_duplicates(subset='PassengerId')
```

```python
df.head()
```

2. Backfill missing values in the Age column using the median of the existing ages.

```python
# Calculate the median of the 'Age' column
median_age = df['Age'].median()
```

```python
# Backfill missing values in the 'Age' column with the median value
df['Age'].fillna(median_age, inplace=True)
```

```python
df.head()
```

3. Create a scatter plot with the Fare on the x-axis and Age on the y-axis. This will help you visualize the relationship between passenger fare and age.

```python
#Creating a scatter plot with Fare on the x-axis and Age on the y-axis
plt.figure(figsize=(10, 6))
plt.scatter(df['Fare'], df['Age'], alpha=0.5, color='b')
plt.title('Scatter Plot of Fare vs Age', fontsize=14)
plt.xlabel('Fare', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.grid(True)
plt.show()
```

4. Calculate and display the survival rate for each passenger class (Pclass).

```python
#Group by 'Pclass' and calculate the mean survival rate for each class
class_survival_rate = df.groupby('Pclass')['Survived'].mean()

# Display the survival rates for each passenger class
print(class_survival_rate)
```

5. Calculate and display the survival rate for each gender.

```python
# Group by 'Sex' and calculate the mean survival rate for each gender
gender_survival_rate = df.groupby('Sex')['Survived'].mean()

# Display the survival rates for each gender
print(gender_survival_rate)
```
